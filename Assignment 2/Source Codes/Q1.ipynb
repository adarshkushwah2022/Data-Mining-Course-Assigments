4.Visualizing Data

```python
#finding most frequent userId
l1=dict(df1['userId'].value_counts())
Keymax1 = max(zip(l1.keys(),l1.values()))
print("The most frequent userID is",Keymax1[0],"and its frequency is",Keymax1[1])
```

```python
#Evaluating and storing frquency of all ratings given by all users
ratings_freq=list()
rating_values=[0.5,1,1.5,2,2.5,3,3.5,4,4.5,5]
df_1=list(df1['rating'])
for value in rating_values:
  count=0
  for rating in df_1:
    if(value==rating):
      count=count+1
  ratings_freq.append(count)
```

```python
#Pie chart to visualize the frequency of ratings given by all the users
fig = plt.figure(figsize =(10, 7))
plt.pie(ratings_freq, labels = rating_values)
plt.title("Frequency of Ratings used by all the users")
plt.show()
```

```python
#plotting correlation graphs
sns.pairplot(df1,hue="rating",height=4)
```

```python
#Finding which user tagged the most number of movies
unique_userIds=list(df2.userId.unique())
most_rated_user=-1
number_of_movies_rated_most=-1
x=list()
y=list()
for ele in unique_userIds:
  count=0
  for cur_row in df2.iterrows():
    if(cur_row[1][0]==ele):
      count=count+1
  if(count>number_of_movies_rated_most):
    most_rated_user=ele
    number_of_movies_rated_most=count
  x.append(ele)
  y.append(count)


plt.xlabel('UserId')
plt.ylabel('Number Of Movies Tagged')
plt.plot(x, y)
print("User with the userId:",most_rated_user,"rated the highest number of movies:",number_of_movies_rated_most)
```

2.Statistical Insights

3.Data Cleaning

```python
#The number of missing values for each variable
df3.isnull().sum 
```

```python
#Finding total number of duplicates entries
df3.duplicated().sum()
```

4.Data Visualization

```python
#Pie Chart
labels=df3['genres'].unique().tolist()
labels_list = list(range(0, len(labels)))
slices =[]
df_3=list(df3['genres'])
for l in labels:
    count=0
    for genres in df_3:
        if(l==genres):
           count=count+1
    slices.append(count)
```

```python
fig = plt.figure(figsize =(20, 10))
plt.pie(slices)
plt.title("Distribution of movies genre wise")
plt.show()
```

```python
#Bar Plot
x_val = labels_list

y_val = slices

plt.axis([0, 900, 0, 210])

plt.bar(x_val, y_val)

plt.xlabel('Genres')
plt.ylabel('Count of movies')
plt.title('Count of movies per genre')
```
